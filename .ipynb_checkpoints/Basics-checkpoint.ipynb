### Creating helper functions
In order to more easily access our sqlite database we'll first create some helper functions that will
* take in a SQL query string
* connect to the database
* exceute the SQL query

```python
#database name on file
db = 'chinook.db'

def run_query(q):
#takes SQL query as argument, returns pandas dataframe of query result
    with sqlite3.connect(db) as conn:
        return pd.read_sql(q, conn)
    
def run_command(c):
#takes SQL command, executes using sqlite module    
    with sqlite3.connect(db) as conn:
        conn.isolation_level = None
        conn.execute(c)
        
def show_tables():
#calls run_query to return a list of all tables and views in databse
    query_string = """
    SELECT
        name,
        type
    FROM sqlite_master
    WHERE type IN ("table","view");    
    """
    return run_query(query_string)

show_tables()
```

## US Track Sales by Genre

We are tasked with selecting three albums from the four listed below to be added to the store. 

|Artist Name|Genre|
|---|---|
|Regal|Hip-Hop|
|Red Tone|Punk|
|Meteor and the Girls|Pop|
|Slim Jim Bites|Blues|

The record label we're getting these albums will provide money to advertise the new albums in the US, so, our goal is to find out which genres have the highest sales in the US.

### SQL Strategy

write a query that returns:
- names of genres
- respective sales for reach genre (from US customers)

```python
q = """
WITH us_track_sales AS 
    (SELECT il.* 
        FROM invoice_line il
        INNER JOIN invoice i on i.invoice_id = il.invoice_id
        INNER JOIN customer c on c.customer_id = i.customer_id
        WHERE c.country = "USA")

SELECT
    g.name genre,
    SUM(uts.quantity) tracks_sold,
    (CAST(SUM(uts.quantity) AS FLOAT) / (
        SELECT COUNT(*) FROM us_track_sales
    ))percentage_sold
FROM us_track_sales uts
INNER JOIN track t on t.track_id = uts.track_id
INNER JOIN genre g on g.genre_id = t.genre_id
GROUP BY 1
ORDER BY 2 DESC
LIMIT 10

"""

run_query(q)
```

```python
#visualize results shown above
genre_sales_usa = run_query(q)

#set genre as index to ease data viz labeling
genre_sales_usa.set_index('genre', inplace=True, drop=True)

#create barplot of results
genre_sales_usa['tracks_sold'].plot.barh(
    xlim=(0,650), #adjust x-axis range so percentage fits on graph
    colormap=plt.cm.Accent) #colormap presents preset colors graph

#Title/Label adjustments
plt.title('Top Selling Genres in US',fontsize=16)
plt.ylabel('')


#add respective perecentages on ends of bars
for i, label in enumerate(list(genre_sales_usa.index)):
    score = genre_sales_usa.loc[label, "tracks_sold"]
    label = (genre_sales_usa.loc[label, "percentage_sold"] * 100
            ).astype(int).astype(str) + "%"
    plt.annotate(str(label), (score + 10 , i - 0.15))

plt.show()
```

### Analysis
From the above graph we can clearly see that Rock, Punk, and Metal are the most popular genre types in the USA, with Blues, Pop and Hip-Hop/Rap finishing off the list, the latter two being approximately equal in market share(2.1% Pop, 1.9% Hip-Hop). 

We suggest signing the three most popular genre artists available: Red Tone, Slim Jim Bites, and Meteor and the Girls. We also suggest looking for signing on more Rock groups in the future, as these are by far the most popular on our platform

## Sales Support Agent Performance
We're tasked with analyzing the performance of our employees in Sales Support. Each customer is assigned a sales support agent, we would like to investigate any variance we see, and whether that variance might be indicative of employee performance.

```python
#show customer sales data (used as a view for tabulating sales data) 
customer_totals="""
SELECT 
    c.customer_id, 
    COUNT(i.invoice_id) number_of_purchases,
    SUM(i.total) total_sales_customer,
    c.country,
    c.support_rep_id
FROM customer C
INNER JOIN invoice i ON c.customer_id = i.customer_id
GROUP BY 1 
ORDER BY 2 DESC
"""
run_query(customer_totals)
```

```python
#display aggregated sales data from customer/employee tables
q="""
WITH customer_info as
(SELECT 
    c.customer_id, 
    COUNT(i.invoice_id) number_of_purchases,
    SUM(i.total) total_sales,
    c.country,
    c.support_rep_id
FROM customer C
INNER JOIN invoice i ON c.customer_id = i.customer_id
GROUP BY 1
)

SELECT 
    (e.first_name || " " || e.last_name) employee,
    SUM(total_sales) total_sales,
    e.hire_date,
    COUNT(customer_id) num_cust,
    AVG(total_sales) sale_per_cust,
    COUNT(DISTINCT ci.country) num_of_countries 
FROM customer_info ci
INNER JOIN employee e ON e.employee_id = ci.support_rep_id
GROUP BY 1
ORDER BY 2 DESC
"""

run_query(q)
```

```python
#sales data viz
employee_sales = run_query(q)
employee_sales.set_index('employee', inplace=True, drop=True)

#order sales by total sales
employee_sales.sort_values("total_sales",inplace=True)

#create barplot
employee_sales['total_sales'].plot.barh(colormap=plt.cm.Accent)

#titles/fonts
plt.title("Sales per employee",fontsize=14)    
plt.ylabel('')
plt.xlabel('Total Sales (USD)',fontsize=12)

#display # of customers inside bars
for i, label in enumerate(list(employee_sales.index)):
    score = employee_sales.loc[label,"total_sales"]
    label = (employee_sales.loc[label, "num_cust"]).astype(str) + " customers"
    plt.annotate(label, (score - 400, i))

plt.show()
```

### Sales correlate with number of customers, time at company
Jane Peacock has the highest sales of any support reps. However, looking at the table, each employee's sale totals roughly correlate to their hire dates, which is further reflected in the total number of customers each employee is responsible for. And while Steve Johnson has the fewest total sales, he is also responsible for customers in more countries than either other support rep, the language barriers that brings may also contribute to his lower sales/customer metric. We believe that each employee is performing their duties adequately

## Customer Purchasing Trends by Country

We've been tasked with analyzing the sales data for customers from different countries. Particularly, for each country, we need:
- total number of customers
- total value of sales
- average value of sales per customer
- average order value

```python
q="""
WITH country_or_other AS
    (SELECT
        CASE
            WHEN (
                 SELECT count(*)
                 FROM customer
                 where country = c.country
                ) = 1 THEN "Other"
           ELSE c.country
        END AS country,
        c.customer_id,
        i.invoice_id,
        i.total
    FROM customer c
    INNER JOIN invoice i on c.customer_id = i.customer_id
    )

SELECT
    country,
    customers,
    total_sales,
    avg_order,
    customer_lifetime_value
FROM (
    SELECT
        CASE
            WHEN country = "Other" THEN 1
            ELSE 0 
        END AS sort, 
        country,
        count(distinct customer_id) customers,
        SUM(total) total_sales,
        (SUM(total) / count(distinct customer_id)) customer_lifetime_value,
        (SUM(total) / count(invoice_id)) avg_order
    FROM country_or_other
    GROUP BY country
    ORDER BY sort ASC, total_sales DESC
    )
"""
run_query(q)
```

```python
#create data viz for above table
sales_data = run_query(q)
sales_data = sales_data.set_index('country')
labels = list(sales_data.index)

#select colors for different countries to be used
colors = [cm.Paired(i) for i in np.linspace(0,1,sales_data.shape[0])]

#create plot, format 4x4
fig, ax = plt.subplots(nrows=2,ncols=2, figsize=(10,10))
ax1, ax2, ax3, ax4 = ax.flatten()
fig.subplots_adjust(hspace=.5)

#topleft: total sales by country
sales_breakdown = sales_data['total_sales'].copy().rename('') #used to remove xlabel ('total sales') from pie chart
sales_breakdown.plot.pie(labels=labels,
                                   ax=ax1,
                                   colors=colors,
                                   startangle=-90,)
ax1.set_title('Total Sales by Country')
ax1.set_aspect("equal")


#topright: customer lifetime value
customer_val = sales_data['customer_lifetime_value'].copy().rename("")
customer_val.index.set_names('', inplace=True)
customer_val.drop('Other',inplace=True)
customer_val.plot.bar(ax=ax2,
                      color=colors, 
                      label='avg_order_per_customer')
ax2.set_title("Customer Lifetime Value, Dollars")
ax2.tick_params(top="off", right="off", left="off", bottom="off")

#botleft: avg_order value, number of customers
orders_customers = sales_data[['customers', 'avg_order']].copy()
orders_customers.rename({'avg_order':'avg_order (US)'}, inplace=True)
orders_customers.index.name= ''
orders_customers.drop('Other', inplace=True)
orders_customers.plot.bar(ax=ax3,
                     colormap = cm.Set1,
                     label='thisisalabel')
ax3.set_title("Number of Customers vs Avg Order Value (USD)")

plt.show()
```

USA currently has the largest market share of our business in both customers and revenue, and making decisions around their buying habits is reasonable. For new and emerging markets, the Czech Republic and India have some of the fewest customers, but relatively high average order values. 

```python
#invoices containing tracks from an album
q="""

WITH in_al AS  

(SELECT
    il.invoice_id,
    t.album_id
FROM invoice_line il
INNER JOIN track t ON il.track_id = t.track_id
WHERE t.album_id IN (SELECT album_id
                     FROM track t
                     GROUP BY 1
                     HAVING COUNT(track_id) > 2) 
GROUP BY 1
)

Select * from in_al
"""

run_query(q)
```

## Album vs Track Purchasing Strategy

In order to maximize revenue, management is currently considering changing their purchasing strategy from the purchase of full albums, to the purchase of the most popular tracks from each album.

**Find out what percentage of purchases are individual tracks vs whole albums**


```python
#all tracks purchased on invoice_ids
q="""
WITH sq1 AS 
(SELECT
    il.invoice_id,
    il.track_id,
    t.album_id
FROM invoice_line il
INNER JOIN track t ON il.track_id = t.track_id)


SELECT
    SUM(full_album_purchase) album_invoices,
    COUNT(*) - SUM(full_album_purchase) non_album_invoices,
    AVG(full_album_purchase) album_purchases,
    (1 - AVG(full_album_purchase)) non_album_purchases
FROM 
(SELECT
    CASE
    WHEN
        (
        SELECT track_id FROM track 
        WHERE album_id = s.album_id
        EXCEPT
        SELECT track_id FROM sq1
        WHERE invoice_id = s.invoice_id
        ) IS NULL
        
        AND
        
        (SELECT track_id FROM sq1
        WHERE invoice_id = s.invoice_id
        EXCEPT
        SELECT track_id FROM track 
        WHERE album_id = s.album_id
        ) IS NULL 
        
        
        THEN 1
        ELSE 0
    END
    AS full_album_purchase        
FROM sq1 s
GROUP BY s.invoice_id)

"""


run_query(q)
```

From the above query, we see that albums account for 18.5% of our purchases. Based on this data, we advise against changing our buying scheme as there's the potential to lose a fifth of our revenue.
